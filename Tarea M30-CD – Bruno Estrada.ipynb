# Problema NEAREST NEIGHBORS

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
vino = pd.read_csv('wine-clustering.csv')
vino.head()
```

```python
test = pd.DataFrame({
    'Alcohol': [14],
    'Malic_Acid': [2],
    'Ash': [2.5],
    'Ash_Alcanity': [16],
    'Magnesium': [115],
    'Total_Phenols': [3],
    'Flavanoids': [2.5],
    'Nonflavanoid_Phenols': [0.4],
    'Proanthocyanins': [2],
    'Color_Intensity': [9],
    'Hue': [1],
    'OD280': [3.5],
    'Proline': [800]
})

test
```

```python
KNN_modelo_regresion = NearestNeighbors(n_neighbors = 5, algorithm='auto')
KNN_modelo_regresion.fit(vino)
```

```python
# Encontrar los vecinos más cercanos
distancias, indices = KNN_modelo_regresion.kneighbors(test)

# Imprimir los resultados
print("Distancias a los vecinos más cercanos:", distancias)
print("Índices de los vecinos más cercanos:", indices, "\n")

# Mostrar los vinos más cercanos
vinos_mas_cercanos = pd.DataFrame(vino.iloc[indices[0]])
print("Los 5 vinos más cercanos:")
vinos_mas_cercanos
```

# Problema MARKET BASQUET

```python
mi_canasta_clientes = [['bread', 'butter', 'wine', 'bananas', 'coffee', 'carrots'],
    ['tomatoes', 'onions', 'cheese', 'milk', 'potatoes'],
    ['beer', 'chips', 'asparagus', 'salsa', 'milk', 'apples'],
    ['olive oil', 'bread', 'butter', 'tomatoes', 'steak', 'carrots'],
    ['tomatoes', 'onions', 'chips', 'wine', 'ketchup', 'orange juice'],
    ['bread', 'butter', 'beer', 'chips', 'milk'],
    ['butter', 'tomatoes', 'carrots', 'coffee', 'sugar'],
    ['tomatoes', 'onions', 'cheese', 'milk', 'potatoes'],
    ['bread', 'butter', 'ketchup', 'coffee', 'chicken wings'],
    ['butter', 'beer', 'chips', 'asparagus', 'apples'],
    ['tomatoes', 'onion', 'beer', 'chips', 'milk', 'coffee']]
```

```python
import pandas as pd
import numpy as np

# Crear el DataFrame inicial
todos_los_productos = list(set(product for basket in mi_canasta_clientes for product in basket))
clientes = range(len(mi_canasta_clientes))

Tabla_clientes = pd.DataFrame(
    data=np.zeros((len(clientes), len(todos_los_productos)), dtype=int),
    columns=todos_los_productos,
    index=clientes
)

# Rellenar la tabla con un bucle
for cliente, productos in enumerate(mi_canasta_clientes):
    for producto in productos:
        Tabla_clientes.at[cliente, producto] = 1

# Mostrar la tabla resultante
Tabla_clientes
```

### Revision extra para revisar integridad de los datos

```python
# Bucle para cada cliente
for cliente in Tabla_clientes.index:  # Recorre cada índice del DataFrame (clientes)
    productos_comprados = Tabla_clientes.loc[cliente][Tabla_clientes.loc[cliente] == 1].index.tolist()
    print(f"Cliente {cliente} compró: {productos_comprados}")
    print(f"Los productos están íntegros: {set(productos_comprados) == set(mi_canasta_clientes[cliente])} ")
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
compras_frecuentes = apriori(Tabla_clientes, min_support= 0.3, 
                             use_colnames = True ).sort_values('support', ascending = False).reset_index(drop=True)
compras_frecuentes['length']= compras_frecuentes['itemsets'].apply(lambda x: len(x))
compras_frecuentes
```

## Explicacion:

### En los datos aparece mantequilla y tomates más del 50% quiere decir que lo compran recurrentemente, y que el dos veces 36% aparece el conjunto (cerveza, papas) y (mantequilla, pan)

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(compras_frecuentes, metric= 'confidence', 
                             min_threshold = 0.5 ).sort_values('confidence', ascending = False).reset_index(drop=True)
```

## Explicacion:

### El 100% de los casos donde compran cerveza tambien papas, e igual el 100% de las veces que se compra pan, se lleva mantequilla. El 80% si es viceversa, si se compran papas se lleva cerveza y si se lleva mantequilla se llevan pan pero con 66% de probabilidad.

### Es 2.2 veces más probable que compren papas si cerveza es comprada, mientras que en el pan es 1.8 veces más probable que se lleven mantequilla si llevan pan.
